```python
## generating parental column

choices = ['Yes Yes', 'Yes No','No Yes', 'No No']

parental_mother_grandmother = []
parental_father_grandfather = []

for i in range(0, 100000):
    temp1 = random.choice(choices)
    temp2 = random.choice(choices)
    parental_mother_grandmother.append(temp1)
    parental_father_grandfather.append(temp2)
```

```python
choices_daily_Activity = ['walking', 'yoga', 'swimming', 'dancing','gym']

daily_activity = []

for i in range(0,100000):
    number = random.randint(1,5)
    temp = random.choices(choices_daily_Activity, k=number)
    temp = list(set(temp))
    temp = ' '.join(temp)
    daily_activity.append(temp)
```

```python
choice_gender = ['Male', 'Female']

age = []
gender = []

for i in range(0,100000):
    number_age = random.randint(10,90)
    rand_gender = random.choice(choice_gender)
    age.append(number_age)
    gender.append(rand_gender)
    
```

```python
choice_education_level = ['undergrade','graduated','below undergrade','None']
choice_race = ['White','Black or African American', 'Asian', 'American Indian or Alaska Native', 'Native Hawaiian or Pacific Islander']
education_level = []
race = []

for i in range(0,100000):
    temp = random.choice(choice_education_level)
    temp2 = random.choice(choice_race)
    education_level.append(temp)
    race.append(temp2)
```

```python
taking_prescription_hypertension = []
choice_hypertension = ['Yes', 'No', 'Not Known', '']

choice_AMD = ['none','intermediate','late', '']
AMD_stage = []

choice_presence_any_AMD = ['Yes', 'No', 'Not Known', '']
presence_any_AMD = []

for i in range(0,100000):
    temp = random.choice(choice_hypertension)
    temp2 = random.choice(choice_AMD)
    temp3 = random.choice(choice_presence_any_AMD)
    
    taking_prescription_hypertension.append(temp)
    AMD_stage.append(temp2)
    presence_any_AMD.append(temp3)
    
    
```

```python
choices = ['Yes','No','Refused',"Don't know",'Missing']

Are_you_reducing_fat_or_calories_in_diet = []
Are_you_controlling_weight = []
Are_you_increasing_physical_activity = []

Are_you_reducing_fat_or_calories_in_diet_past_three_year = []
Are_you_controlling_weight_past_three_year = []
Are_you_increasing_physical_activity_past_three_year = []

Ever_told_have_health_risk_for_diabetes = []
Ever_told_you_have_prediabetes = []

Had_blood_tested_past_three_years = []
Diabetes_affected_eyes_had_retinopathy = []

for i in range(0,100000):
    temp = random.choice(choices)
    temp1 = random.choice(choices)
    temp2 = random.choice(choices)
    temp3 = random.choice(choices)
    temp4 = random.choice(choices)
    temp5 = random.choice(choices)
    temp6 = random.choice(choices)
    temp7 = random.choice(choices)
    temp8 = random.choice(choices)
    temp9 = random.choice(choices)
    
    
    
    Are_you_reducing_fat_or_calories_in_diet.append(temp)
    Are_you_controlling_weight.append(temp1)
    Are_you_increasing_physical_activity.append(temp2)
    Are_you_reducing_fat_or_calories_in_diet_past_three_year.append(temp3)
    Are_you_controlling_weight_past_three_year.append(temp4)
    Are_you_increasing_physical_activity_past_three_year.append(temp5)
    Ever_told_have_health_risk_for_diabetes.append(temp6)
    Ever_told_you_have_prediabetes.append(temp7)
    Had_blood_tested_past_three_years.append(temp8)
    Diabetes_affected_eyes_had_retinopathy.append(temp9)
```

```python

data = {
    'sugar_level' : sugar_level,
    'parental_mother_grandmother' : parental_mother_grandmother,
    'parental_father_grandfather' : parental_father_grandfather,
    'daily_activity' : daily_activity,
    'age' : age,
    'gender' : gender,
    'education_level' : education_level,
    'race' : race,
    'taking_prescription_hypertension':taking_prescription_hypertension,
    'presence_any_AMD':presence_any_AMD,
    'AMD_stage':AMD_stage,
    'Are_you_reducing_fat_or_calories_in_diet':Are_you_reducing_fat_or_calories_in_diet,
    'Are_you_controlling_weight':Are_you_controlling_weight,
    'Are_you_increasing_physical_activity':Are_you_increasing_physical_activity,
    'Are_you_reducing_fat_or_calories_in_diet_past_three_year':Are_you_reducing_fat_or_calories_in_diet_past_three_year,
    'Are_you_controlling_weight_past_three_year':Are_you_controlling_weight_past_three_year,
    'Are_you_increasing_physical_activity_past_three_year':Are_you_increasing_physical_activity_past_three_year,
    'Ever_told_have_health_risk_for_diabetes':Ever_told_have_health_risk_for_diabetes,
    'Ever_told_you_have_prediabetes':Ever_told_you_have_prediabetes,
    'Had_blood_tested_past_three_years':Had_blood_tested_past_three_years,
    'Diabetes_affected_eyes_had_retinopathy':Diabetes_affected_eyes_had_retinopathy
}

dataset_diabetes_detection = pd.DataFrame(data)
```

```python
## for response variable


median_sugar = dataset_diabetes_detection['sugar_level'].median()
```

```python
median_sugar
occurance_of_diabetes = []

sugar_level = dataset_diabetes_detection['sugar_level']

for i in range(0,100000):
    if sugar_level[i] > median_sugar:
        occurance_of_diabetes.append('Yes')
    else:
        occurance_of_diabetes.append('No')
```

```python
dataset_diabetes_detection.insert(0,"occurance_of_diabetes", occurance_of_diabetes,True)
```
